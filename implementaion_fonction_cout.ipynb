```python
import random
```

```python
x = list(range(0,55,3))
x
```

```python
y = list(i*2 for i in x)
y
```

```python
w=1
b=0
```

```python
def model(x ,w ,b) :
  return w*x + b
```

```python
def cost(y , y_pred) :
  sum = 0
  for i in range(len(y)):
    sum += (y[i] - y_pred[i])**2
  return sum/(2*len(y))
```

```python
def deriv_w(x , y , y_pred) :
  dw = 0
  for i in range(len(y)) :
    dw += ((y_pred[i] - y[i]) * x[i])
  return dw/len(y)
```

```python
def deriv_b(y , y_pred) :
  db = 0
  for i in range(len(y)) :
    db += ((y_pred[i] - y[i]))
  return db/len(y)
```

```python
def update(w , b , x , y , y_pred , learning_rate= 0.001) :
  w -=  learning_rate*deriv_w(x ,y , y_pred)
  b -=  learning_rate*deriv_b(y , y_pred)
  return w , b

```

```python
list_cout = list()
n_iter = 100
for i in range(n_iter):
  y_pred = list(model(x_i , w , b) for x_i in x)
  c = cost(y , y_pred)
  list_cout.append(c)
  w , b = update(w , b ,x ,y ,y_pred)
w_final = w
b_final = b
```

```python
list_cout
```

```python
import matplotlib.pyplot as plt

plt.plot(list_cout)
```

```python
import matplotlib.pyplot as plt

# 1. On affiche les données d'entraînement (les points)
plt.scatter(x, y, color='blue', label='Données réelles')

# 2. On calcule les points de la droite avec les paramètres w et b trouvés
# y_final = w * x + b
y_final = [w_final * x_i + b_final for x_i in x]

# 3. On trace la droite
plt.plot(x, y_final, color='red', linewidth=2, label='Modèle (droite de régression)')

# 4. Habillage du graphique
plt.title(f"Prix des maisons en fonction de la surface (w={w:.2f}, b={b:.2f})")
plt.xlabel("Surface (m²)")
plt.ylabel("Prix (€)")
plt.legend()
plt.show()
```
